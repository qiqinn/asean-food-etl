```python
from pyspark.sql import SparkSession

# Start Spark
spark = SparkSession.builder \
    .appName("WFP Food Prices ETL") \
    .getOrCreate()

# Step 2: Load raw data
input_path = "../data/raw/wfp_food_prices_global_2025.csv"
# df = spark.read.option("header", True).option("inferSchema", True).csv(input_path)
```

```python

df = spark.read.parquet("/home/qin/projects/asean-food-etl/data/processed/asean_food.parquet/*")


# Step 3: Show schema and sample rows   
print("=== Schema ===")
df.printSchema()
```

```python

print("=== Sample Data ===")
df.limit(5).toPandas()


# Step 5: Stop Spark session
```
